```python
import cv2
import face_recognition
import pickle
import os
```

```python
# Importing images of Students
ImagesPath = 'Images' # relative to this notebook
pathList = os.listdir(ImagesPath)
print(pathList)
imgList = []
studentIds = []
for path in pathList:
    imgPath = os.path.join(ImagesPath, path)  
    imgList.append(cv2.imread(imgPath))
    studentIds.append(os.path.splitext(path)[0])
print(studentIds)
```

```python
print(imgList)
```

```python
def findEncodings(imagesList):
    encodeList = []
    for img in imagesList:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
        encode = face_recognition.face_encodings(img)[0]
        encodeList.append(encode)

    return encodeList
```

OpenCV uses BGR color format but face_recognition library uses RGB. Therefore we need to convert BGR to RGB using -> img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

```python
# generate encoding
print("Encoding Started ...")
encodeListKnown = findEncodings(imgList)
encodeListKnownWithIds = [encodeListKnown, studentIds]
print("Encoding Complete")
```

```python
# saving encoding file using pickle
file = open("EncodeFile.p", 'wb') # wb for write mode
pickle.dump(encodeListKnownWithIds, file)
file.close()
print("File Saved")
```
